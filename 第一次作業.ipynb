```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 畫函數圖形
畫 $f(x) = \sin(x)$

```python
x = np.linspace(0, 2 * np.pi, 500)
y = np.sin(x)

plt.plot(x, y, linewidth=3) # Increased linewidth
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('y = sin(x)')
plt.grid(True)
plt.show()
```
