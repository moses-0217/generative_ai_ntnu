```python
%matplotlib inline

#引入函式庫
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 畫函數圖形
畫 $f(x) = \tanh(x)$

```python
x = np.linspace(-4, 4, 500) #x值範圍從-4到4，共500個點
y = np.tanh(x) #tanh(x)函數

plt.plot(x, y, color='orange', linewidth=3) #繪製圖形，設定線條顏色為橘色並增加線寬
plt.xlabel('x') #x軸標籤
plt.ylabel('tanh(x)') #y軸標籤
plt.title('y = tanh(x)') #設定圖表標題
plt.grid(True) #顯示網格
plt.show() # 顯示圖表
```
